```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib.ticker import FuncFormatter

data = {
    "State_Name": ["Iowa", "Pennsylvania", "Texas", "Ohio", "Minnesota"],
    "Total_Milk_Production": [
        457069400000,
        394351200000,
        288155700000,
        285967500000,
        273658700000,
    ],
}

df = pd.DataFrame(data)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

ax = sns.barplot(
    x="Total_Milk_Production",
    y="State_Name",
    data=df,
    hue="State_Name",
    palette="Blues_r", 
    legend=False
)

plt.title(
    "Top 5 U.S. States by Total Milk Production (USDA Data)",
    fontsize=14,
    weight="bold",
)

def format_billion(x, pos):
    return f"{x*1e-9:.0f}B"

ax.xaxis.set_major_formatter(FuncFormatter(format_billion))

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.savefig("top_5_milk_producers.png", dpi=300, bbox_inches="tight")
plt.show()
```
