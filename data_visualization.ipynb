```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Menyiapkan data hasil kueri SQL Anda sebelumnya
data = {
    'Nama_Negara_Bagian': ['NEW YORK', 'CALIFORNIA', 'PENNSYLVANIA', 'TENNESSEE', 'WISCONSIN'],
    'Total_Yogurt': [10919.0, 10262.0, 2428.0, 1601.0, 1402.0]
}

# 2. Mengubah data menjadi DataFrame Pandas
df = pd.DataFrame(data)

# 3. Mengatur tema gaya visual dari Seaborn
sns.set_theme(style="whitegrid")

# 4. Membuat ukuran kanvas grafik (Lebar 9 inci, Tinggi 5 inci)
plt.figure(figsize=(9, 5))

# 5. Membuat grafik batang menggunakan Seaborn dengan palet warna menarik
# Kita gunakan grafik horizontal (x=angka, y=kategori) agar teks nama wilayah mudah dibaca
ax = sns.barplot(
    x='Total_Yogurt', 
    y='Nama_Negara_Bagian', 
    data=df, 
    palette='viridis',  # Palet warna gradasi dari ungu ke kuning kehijauan
    hue='Nama_Negara_Bagian', # Mengaktifkan pewarnaan per bar
    legend=False
)

# 6. Memberikan judul grafik dan nama sumbu (Labeling)
plt.title('Top 5 Yogurt Producing States in the US', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Total Production Volume', fontsize=12, labelpad=10)
plt.ylabel('State Name', fontsize=12)

# 7. Menambahkan label angka di ujung setiap batang secara otomatis
for container in ax.containers:
    ax.bar_label(container, fmt='%,.0f', padding=5, fontsize=10, fontweight='bold')

# 8. Menampilkan grafik secara rapi
plt.tight_layout()
plt.show()
```

```python
plt.savefig('top_5_yogurt_chart.png', dpi=300, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Menyiapkan data madu hasil analisis Anda (Wisconsin, California, New York, Minnesota, Pennsylvania)
data_honey = {
    'Nama_Negara_Bagian': ['WISCONSIN', 'CALIFORNIA', 'NEW YORK', 'MINNESOTA', 'PENNSYLVANIA'],
    'Total_Madu': [3788649000000, 3362642000000, 2019729000000, 1743875000000, 1511790000000]
}

# 2. Mengubah menjadi DataFrame Pandas
df_honey = pd.DataFrame(data_honey)

# 3. Mengatur gaya visual
sns.set_theme(style="whitegrid")
plt.figure(figsize=(9, 5))

# 4. Membuat grafik horizontal dengan palet warna gradasi jingga-kuning khas madu ('YlOrBr')
ax_honey = sns.barplot(
    x='Total_Madu', 
    y='Nama_Negara_Bagian', 
    data=df_honey, 
    palette='YlOrBr_r',  # Palet warna kuning-oranye-cokelat yang estetik
    hue='Nama_Negara_Bagian',
    legend=False
)

# 5. Memberikan judul dan label sumbu
plt.title('Top 5 Honey Producing States in the US', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Total Honey Production Volume', fontsize=12, labelpad=10)
plt.ylabel('State Name', fontsize=12)

# 6. Menambahkan label angka di samping batang (diformat dengan koma pemisah ribuan)
for container in ax_honey.containers:
    ax_honey.bar_label(container, fmt='%,.0f', padding=5, fontsize=10, fontweight='bold')

# 7. Menyimpan gambar grafik secara otomatis ke folder proyek Anda
plt.savefig('top_5_honey_chart.png', dpi=300, bbox_inches='tight')

# 8. Menampilkan grafik di layar
plt.tight_layout()
plt.show()
```
